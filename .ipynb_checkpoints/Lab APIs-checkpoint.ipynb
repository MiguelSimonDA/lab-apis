```python
import requests
import pandas as pd

poke_all = []

for i in range(1, 152):
    url = f"https://pokeapi.co/api/v2/pokemon/{i}"
    response = requests.get(url)
    response.raise_for_status()
    data = response.json()

    poke_all.append({
        data["name"],
        data["height"],
        data["weight"],
        data["base_experience"],
        data["types"][0]["type"]["name"]
    })

df = pd.DataFrame(poke_all)
df.head()
df.copy()
```

```python


df["weight_kg"] = df["weight"] / 10

heaviest = df.loc[df["weight_kg"] == df["weight_kg"].max()]
print("Heaviest Pokémon:", heaviest[["name", "weight_kg"]])

print("Average weight (kg):", df["weight_kg"].mean())

print("\nPokémon per type:")
print(df["first_type"].value_counts())

print("\nType with highest average weight:")
type_avg = df.groupby("first_type")["weight_kg"].mean()
print(type_avg.loc[type_avg.idxmax()])
```

```python
# Your code goes here
import requests
import pandas as pd

url = "https://api.coingecko.com/api/v3/simple/price"

params = {
    "ids": "bitcoin,ethereum,solana,dogecoin",
    "vs_currencies": "usd"
}

response = requests.get(url, params=params)
response.raise_for_status()
data = response.json()

# Convert to DataFrame (coins as rows)
coins = list(data.keys())
prices = [data[coin]["usd"] for coin in coins]

df = pd.DataFrame({
    "coin": coins,
    "price_usd": prices
})

# Rename columns properly
df.columns = ["coin", "price_usd"]

# Sort by price descending, add rank
df = df.sort_values("price_usd", ascending=False).reset_index(drop=True)
df["price_rank"] = df.index + 1

# Add price category
def get_category(price):
    if price > 1000:
        return "high"
    elif price >= 100:
        return "medium"
    else:
        return "low"

df["price_category"] = df["price_usd"].apply(get_category)

df
```
