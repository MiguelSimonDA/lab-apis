# Challenge 1: Pokémon Dataset & Analysis

## Goal

Build and analyze a Pokémon dataset using the API.

## Example (single request)

Below you can see how to get data for ONE Pokémon:

```python
import requests
import os
import numpy as np
import pandas as pd

url = "https://pokeapi.co/api/v2/pokemon/1"

response = requests.get(url)
data = response.json()

data
```

## Tasks

1. Using the example above, get data for Pokémon IDs from 1 to 151

2. Extract:
   - name
   - height
   - weight
   - base_experience
   - first type

3. Create a DataFrame

4. Add a new column:
   - weight_kg (divide weight by 10)

5. Answer:
   - Which Pokémon is the heaviest?
   - What is the average weight?
   - How many Pokémon are there per type?

6. Answer:
   - Which type has the highest average weight?

```python
data.keys()
```

```python
data['types']
```

```python
# extraction
poke_data = []

for i in range(1,152):
    url = f"https://pokeapi.co/api/v2/pokemon/{i}"
    response = requests.get(url)
    data = response.json()
    
    pokemon_id = data['id']
    name = data['name']
    height = data['height']
    weight = data['weight']
    base_experience = data['base_experience']
    first_type =  data['types'][0]['type']['name']
    single_poke = {'ID': pokemon_id,
                   'name': name,
                   'height': height,
                   'weight': weight,
                   'base_experience': base_experience,
                   'first_type': first_type}
    poke_data.append(single_poke)
```

```python
# creating dataframde
pokemon = pd.DataFrame(poke_data) 
```

```python
pokemon
```

```python
# Add a new column: weight_kg
pokemon['weight_kg'] = (pokemon['weight'])/10
```

```python
pokemon
```

```python
# Which Pokémon is the heaviest?
pokemon.sort_values('weight_kg', ascending=False).head(1)
```

```python
# What is the average weight?
pokemon['weight_kg'].mean()
```

```python
# How many Pokémon are there per type?
pokemon['first_type'].value_counts()
```

```python
# Which type has the highest average weight?
type_avg_weight = pokemon.pivot_table(index='first_type', values='weight_kg', aggfunc='mean')

type_avg_weight
```
