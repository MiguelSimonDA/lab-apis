## Tasks

1. Using the example above, get data for Pokémon IDs from 1 to 151

2. Extract:
   - name
   - height
   - weight
   - base_experience
   - first type

3. Create a DataFrame

4. Add a new column:
   - weight_kg (divide weight by 10)

5. Answer:
   - Which Pokémon is the heaviest?
   - What is the average weight?
   - How many Pokémon are there per type?

6. Answer:

```python

#get data for Pokémon IDs from 1 to 151
import requests

    #IMPORT the data
base_url = "https://pokeapi.co/api/v2/pokemon/"
pokemon_list = []

    # create the list of pokemons
for pokemon_id in range(1, 152):
    url = f"{base_url}{pokemon_id}"
    
    response = requests.get(url)
    data = response.json()
    
    pokemon_list.append(data)
    print(f"ID {pokemon_id}: {data['name'].capitalize()}")

print(f"Total pokemon's number: {len(pokemon_list)} ---")
```

```python
#2. extract the data
for pokemon_id in range(1, 152):
    url = f"{base_url}{pokemon_id}"
    
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        
        # Extract the data:
        name = data["name"].capitalize()
        height = data["height"]
        weight = data["weight"]
        base_exp = data["base_experience"]
        first_type = data["types"][0]["type"]["name"].capitalize()
        
        # save the dictionary
        pokemon_list.append({
            "name": name,
            "height": height,
            "weight": weight,
            "base_experience": base_exp,
            "first type": first_type
        })
        
        print(f"ID {pokemon_id}: {name}")
        print(f"  > Height: {height} | weight: {weight}")
        print(f"  > base_experience: {base_exp}")
        print(f"  > Principal type: {first_type}")
        print("-" * 20)
```

```python
#3. create a data frame
import requests
import pandas as pd

    # set up of the tqable:
pd.set_option('display.max_rows', None)      
pd.set_option('display.max_columns', None)   
pd.set_option('display.width', None)         

base_url = "https://pokeapi.co/api/v2/pokemon/"
pokemon_list = []


for pokemon_id in range(1, 152):
    url = f"{base_url}{pokemon_id}"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        
        pokemon_info = {
            "ID": pokemon_id,
            "Name": data["name"].capitalize(),
            "Height": data["height"],
            "Weight": data["weight"],
            "Base Experience": data["base_experience"],
            "First Type": data["types"][0]["type"]["name"].capitalize()
        }
        pokemon_list.append(pokemon_info)

    # creation of the data frame
df = pd.DataFrame(pokemon_list)


print(df.to_string(index=False)) 

```

```python
#4. Add a new column:
import requests
import pandas as pd

    #set up of the data frame:
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

base_url = "https://pokeapi.co/api/v2/pokemon/"
pokemon_list = []


for pokemon_id in range(1, 152):
    url = f"{base_url}{pokemon_id}"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        
        # define the columns to be diplayed:
        pokemon_info = {
            "ID": pokemon_id,                                            # 1
            "Name": data["name"].capitalize(),                          # 2
            "Height": data["height"],                                   # 3
            "Weight": data["weight"],                                   # 4
            "Base Experience": data["base_experience"],                 # 5
            "First Type": data["types"][0]["type"]["name"].capitalize(),# 6
            "Weight_kg": data["weight"] / 10                            # 7
        }
        pokemon_list.append(pokemon_info)

# Create the dataframe
df = pd.DataFrame(pokemon_list)


print(df.to_string(index=False))
```

```python
#5.


#Which Pokémon is the heaviest?
heaviest_pokemon = df.loc[df['Weight_kg'].idxmax()]
#What is the average weight?
average_weight = df['Weight_kg'].mean()
#How many Pokémon are there per type?
pokemon_per_type = df['First Type'].value_counts()

print(f"the heaviest pokemon is: {heaviest_pokemon['Name']} con {heaviest_pokemon['Weight_kg']} kg.")
print(f"the average weight is: {average_weight:.2f} kg.")
print("Pokemon per type:")
print(pokemon_per_type)
```

```python
#6. Which type has the highest average weight?

# calculate the average:
type_averages = df.groupby('First Type')['Weight_kg'].mean()

# get the highest one:
top_1_type = type_averages.idxmax()
top_1_value = type_averages.max()

print(f"highest average weight: {top_1_type}")
print(f"average weight: {top_1_value:.2f} kg")
```

```python
# Get prices for: bitcoin, ethereum, solana, dogecoin

import requests
import pandas as pd

url = "https://api.coingecko.com/api/v3/simple/price"

params = {
    'ids': 'bitcoin,ethereum,solana,dogecoin',
    'vs_currencies': 'usd'
}

response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    
    # Convert the dictionary in dataframe
    df_prices = pd.DataFrame(data).T
    
  
    df_prices.columns = ['Price (USD)']
    
   
    df_prices.index = [coin.capitalize() for coin in df_prices.index]
    
    print("Prices")
    print(df_prices)
else:
    print(f"Error : {response.status_code}")
```

```python
#2.Convert the response into a DataFrame (coins as rows)
#convert the data frama
df = pd.DataFrame(data).T

#clean the data frame
df.index.name = 'Coin'           
df.columns = ['Price_USD']       
df.index = df.index.str.capitalize() 


print(df)
```

```python
#3. Rename columns properly
# Rename the columns:
df = df.rename(columns={'usd': 'Price_USD'})

df.index.name = 'Coin_Name'

print(df)
```

```python
#4. Add:price_rank (ranking by price), price_category: high, medium, low
# convert the dataframe
df = pd.DataFrame(data).T
df.index.name = 'Coin'
df.index = df.index.str.capitalize()

# rename the column price
df = df.rename(columns={'usd': 'Price_USD'})

# new columns

    #add price rank
df['price_rank'] = df['Price_USD'].rank(ascending=False).astype(int)

    #price category
def categorize_price(price):
    if price > 1000:
        return "high"
    elif price >= 100:
        return "medium"
    else:
        return "low"

df['price_category'] = df['Price_USD'].apply(categorize_price)


print(df.sort_values('price_rank'))
```

```python
#5
    #Which coin is the most expensive?
most_expensive_coin = df['Price_USD'].idxmax()
highest_price = df['Price_USD'].max()
    #Which category has more coins?
most_frequent_category = df['price_category'].value_counts().idxmax()
category_counts = df['price_category'].value_counts()

print(f"The coin most expensive: {most_expensive_coin} (${highest_price:,.2f})")
print(f"The category most expensive: '{most_frequent_category}'")
```

```python
#6. Create a column with price difference vs bitcoin
# obtain the price of bitcoin
bitcoin_price = df.loc['Bitcoin', 'Price_USD']

# new column
df['diff_vs_bitcoin'] = bitcoin_price - df['Price_USD']


print(df[['Price_USD', 'diff_vs_bitcoin']].sort_values('Price_USD', ascending=False))
```
