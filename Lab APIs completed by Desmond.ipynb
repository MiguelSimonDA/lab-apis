```python
#2 Extract:- name, height, weight, base_experience, first type

extracted = []
base_url = "https://pokeapi.co/api/v2/pokemon/"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}

for i in range(1, 152):
    url = f"{base_url}{i}"
    try:
        resp = requests.get(url, headers=headers, timeout=10)
        resp.raise_for_status()  # Raises an error for 4xx/5xx responses
        data = resp.json()
        extracted.append({
            'name': data['name'],
            'height': data['height'],
            'weight': data['weight'],
            'base_experience': data['base_experience'],
            'first_type': data['types'][0]['type']['name']
        })
    except Exception as e:
        print(f"Failed on pokemon {i}: {e}")  # Now you'll see the actual error

print(f"Extracted {len(extracted)} pokemon")

```

```python
#3. Create a DataFrame
import pandas as pd
import os

# Create df
df = pd.DataFrame(extracted)

print(df.shape)
print(df.head())
```

```python
#4. Add a new column: weight_kg (divide weight by 10)

df = df.assign(weight_kg=lambda x: x['weight'] / 10)
print(df[['name', 'weight', 'weight_kg']].head())







```

```python
#.5 Answer: Which Pokémon is the heaviest?,What is the average weight?, How many Pokémon are there per type?
# Heaviest Pokémon
heaviest = df.loc[df['weight'].idxmax(), 'name']
print(f"Heaviest Pokémon: {heaviest} ({df['weight'].max() / 10} kg)")

# Average weight
avg_weight = df['weight'].mean() / 10
print(f"Average weight: {avg_weight:.2f} kg")

# Pokémon count per type
type_counts = df['first_type'].value_counts()
print("\nPokémon per type:")
print(type_counts)
```

```python
#6 Answer:   - Which type has the highest average weight?

heaviest_type = df.groupby('first_type')['weight_kg'].mean().idxmax()
avg = df.groupby('first_type')['weight_kg'].mean().max()
print(f"Heaviest type: {heaviest_type} ({avg:.2f} kg avg)")
```

```python
#1. Get prices for:,- bitcoin, ethereum, solana, dogecoin
coins = ['bitcoin', 'ethereum', 'solana', 'dogecoin']
url = "https://api.coingecko.com/api/v3/simple/price"
params = {'ids': ','.join(coins), 'vs_currencies': 'usd'}

resp = requests.get(url, params=params)
data = resp.json()



```

```python
#2. Convert the response into a DataFrame (coins as rows)
df_crypto = pd.DataFrame.from_dict(data, orient='index').reset_index()
print(df_crypto)
```

```python
#3. Rename columns properly
df_crypto.columns = ['coin', 'price_usd']
```

```python
#4. Add:price_rank (ranking by price),price_category,"high" if > 1000,"medium" if between 100 and 1000,"low" if < 100
df_crypto['price_rank'] = df_crypto['price_usd'].rank(ascending=False).astype(int)

df_crypto['price_category'] = pd.cut(
    df_crypto['price_usd'],
    bins=[0, 100, 1000, float('inf')],
    labels=['low', 'medium', 'high']
)
```

```python
#5 Which coin is the most expensive?.Which category has more coins?

most_expensive = df_crypto.loc[df_crypto['price_usd'].idxmax(), 'coin']
print(f"Most expensive: {most_expensive}")
print(f"\nCoins per category:\n{df_crypto['price_category'].value_counts()}")
```

```python
# 6. Bonus: price difference vs bitcoin
btc_price = df_crypto.loc[df_crypto['coin'] == 'bitcoin', 'price_usd'].values[0]
df_crypto['diff_vs_bitcoin'] = df_crypto['price_usd'] - btc_price

print(f"\n{df_crypto.sort_values('price_rank')}")
```
